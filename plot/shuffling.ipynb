```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import brewer2mpl
%matplotlib inline
```

```python
bmap = brewer2mpl.get_map("Set1", "qualitative", 7)                                                                                                                                                                                                                                                                          
colors = bmap.colors
Lcolors = colors[0:5]

for i in range(5):
    color = [i/255. for i in Lcolors[i]]
    Lcolors[i] = color

plt.rcParams['xtick.labelsize']= 17
plt.rcParams['ytick.labelsize']= 17
plt.rcParams["figure.figsize"] = (10,5)
```

```python
synlinear_mean = [0.966, 0.967, 0.98, 0.981, 0.993, 0.998, 1.0]
synlinear_s_mean = [0.549, 0.646, 0.725, 0.81, 0.864, 0.944, 0.988]
heloc_mean = [0.874, 0.883, 0.893, 0.898, 0.906, 0.915, 0.922]
heloc_s_mean = [0.497, 0.52, 0.563, 0.651, 0.766, 0.833, 0.883]
x_xaxis = [5, 10, 20, 40, 80, 160, 320]
```

```python
from matplotlib.ticker import FormatStrFormatter
fig, axs = plt.subplots(1, 2)#, sharex=True, sharey = True)
xtick = np.arange(7)
    
# plot the [2, 0] GMSC

axs[0].plot(xtick, synlinear_mean, color = Lcolors[3], marker = "X", label = "DualCF-without-Shuffling", markerfacecolor = 'none', markersize = 12)
axs[0].plot(xtick, synlinear_s_mean, color = Lcolors[4], marker = "P", label = "DualCF-with-Shuffling", markerfacecolor = 'none', markersize = 12)

 
# plot [2, 1] STD
axs[1].plot(xtick, heloc_mean, color = Lcolors[3], marker = "X", markerfacecolor = 'none', markersize = 12)
axs[1].plot(xtick, heloc_s_mean, color = Lcolors[4], marker = "P", markerfacecolor = 'none', markersize = 12)

axs[0].set_ylabel(r'Agreement$\rightarrow$', fontsize = 17)
axs[0].set_title(r"Syn-Linear", fontsize = 17)
axs[1].set_title(r"Heloc-10", fontsize = 17)

axs[0].set_xticks(xtick)
axs[0].set_xticklabels(x_xaxis)
axs[1].set_xticks(xtick)
axs[1].set_xticklabels(x_xaxis)


axs[0].set_xlabel("#Querys", fontsize = 17)
axs[1].set_xlabel("#Querys", fontsize = 17)
#axs[2, 2].tick_params(axis = 'y', which = 'major', labelsize = 12)
axs[1].yaxis.set_major_formatter(FormatStrFormatter('%.2f'))

handles, labels = axs[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', ncol = 5, fontsize = 15, columnspacing = 2, bbox_to_anchor = (0.5, -0.12))
plt.subplots_adjust(wspace = 0.3)
plt.savefig("shuffling.pdf", dpi = 600, bbox_inches = 'tight')
plt.show()
```
