```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import brewer2mpl
%matplotlib inline
```

```python
synlinear_L2 = [0.966, 0.967, 0.98, 0.981, 0.993, 0.998, 1.0]
synlinear_L1 = [0.953, 0.971, 0.974, 0.979, 0.991, 0.997, 0.999]
synlinear_MAD = [0.968, 0.976, 0.986, 0.989, 0.999, 1.0, 1.0]
synlinear_xaxis = [5, 10, 20, 40, 80, 160, 320]
```

```python
synnonlinear_L2 = [0.864, 0.89, 0.914, 0.939, 0.964, 0.977, 0.992]
synnonlinear_L1 = [0.852, 0.88, 0.897, 0.92, 0.939, 0.964, 0.98]
synnonlinear_MAD = [0.858, 0.883, 0.906, 0.925, 0.939, 0.964, 0.979]
synnonlinear_xaxis = [5, 10, 20, 40, 80, 160, 320]
```

```python
gmsc_L2 = [0.891, 0.905, 0.918, 0.931, 0.945, 0.952, 0.955]
gmsc_L1 = [0.892, 0.902, 0.921, 0.935, 0.943, 0.952, 0.959]
gmsc_MAD = [0.88, 0.901, 0.924, 0.942, 0.949, 0.954, 0.959]
gmsc_xaxis = [5, 10, 20, 40, 80, 160, 320]
```

```python
heloc_L2 = [0.874, 0.883, 0.893, 0.898, 0.906, 0.915, 0.922]
heloc_L1 = [0.837, 0.861, 0.889, 0.906, 0.923, 0.927, 0.937]
heloc_MAD = [0.856, 0.87, 0.894, 0.906, 0.92, 0.929, 0.933]
heloc_xaxis = [5, 10, 20, 40, 80, 160, 320]
```

```python
boston_L2 = [0.836, 0.854, 0.863, 0.87, 0.876, 0.882]
boston_L1 = [0.795, 0.807, 0.844, 0.866, 0.872, 0.886]
boston_MAD = [0.764, 0.806, 0.811, 0.835, 0.852, 0.867]
boston_xaxis = [5, 10, 20, 40, 80, 160]
```

```python
bmap = brewer2mpl.get_map("Set1", "qualitative", 7)                                                                                                                                                                                                                                                                          
colors = bmap.colors
Lcolors = colors[0:5]

for i in range(5):
    color = [i/255. for i in Lcolors[i]]
    Lcolors[i] = color

plt.rcParams['xtick.labelsize']= 17
plt.rcParams['ytick.labelsize']= 17
plt.rcParams["figure.figsize"] = (25, 5)
```

```python
from matplotlib.ticker import FormatStrFormatter
fig, axs = plt.subplots(1,5)#, sharex=True, sharey = True)
xtick = np.arange(7)

axs[0].plot(xtick, synlinear_L2, color = Lcolors[2], label = "L2", marker = "D", markerfacecolor = 'none', markersize = 12)
axs[0].plot(xtick, synlinear_L1, color = Lcolors[0], label = "L1", marker = "s", markerfacecolor = 'none', markersize = 12)
axs[0].plot(xtick, synlinear_MAD, color = Lcolors[1], label = "L1/MAD", marker = "o", markerfacecolor = 'none', markersize = 12)


axs[1].plot(xtick, synnonlinear_L2, color = Lcolors[2], label = "L2", marker = "D", markerfacecolor = 'none', markersize = 12)
axs[1].plot(xtick, synnonlinear_L1, color = Lcolors[0], label = "L1", marker = "s", markerfacecolor = 'none', markersize = 12)
axs[1].plot(xtick, synnonlinear_MAD, color = Lcolors[1], label = "L1/MAD", marker = "o", markerfacecolor = 'none', markersize = 12)

# plot the [2] GMSC
axs[2].plot(xtick, gmsc_L2, color = Lcolors[2], label = "L2", marker = "D", markerfacecolor = 'none', markersize = 12)
axs[2].plot(xtick, gmsc_L1, color = Lcolors[0], label = "L1", marker = "s", markerfacecolor = 'none', markersize = 12)
axs[2].plot(xtick, gmsc_MAD, color = Lcolors[1], label = "L1/MAD", marker = "o", markerfacecolor = 'none', markersize = 12)


# plot the [3] Heloc
axs[3].plot(xtick, heloc_L2, color = Lcolors[2], marker = "D", markerfacecolor = 'none', markersize = 12)
axs[3].plot(xtick, heloc_L1, color = Lcolors[0], marker = "s", markerfacecolor = 'none', markersize = 12)
axs[3].plot(xtick, heloc_MAD, color = Lcolors[1], marker = "o", markerfacecolor = 'none', markersize = 12)

# plot the [4] boston-housing
xtick1 = np.arange(6)
axs[4].plot(xtick1, boston_L2, color = Lcolors[2], marker = "D", markerfacecolor = 'none', markersize = 12)
axs[4].plot(xtick1, boston_L1, color = Lcolors[0], marker = "s", markerfacecolor = 'none', markersize = 12)
axs[4].plot(xtick1, boston_MAD, color = Lcolors[1], marker = "o", markerfacecolor = 'none', markersize = 12)

axs[0].set_ylabel(r'Agreement$\rightarrow$', fontsize = 17)
axs[0].set_title(r'Syn-Linear', fontsize = 17)
axs[1].set_title(r'Syn-Nonlinear', fontsize = 17)
axs[2].set_title(r"GMSC", fontsize = 17)
axs[3].set_title(r"Heloc-10", fontsize = 17)
axs[4].set_title(r"Boston-Housing", fontsize = 17)

axs[0].set_xticks(xtick)
axs[0].set_xticklabels(synlinear_xaxis)
axs[1].set_xticks(xtick)
axs[1].set_xticklabels(synnonlinear_xaxis)
axs[2].set_xticks(xtick)
axs[2].set_xticklabels(gmsc_xaxis)
axs[3].set_xticks(xtick)
axs[3].set_xticklabels(heloc_xaxis)
axs[4].set_xticks(xtick1)
axs[4].set_xticklabels(boston_xaxis)

axs[0].set_xlabel("#Querys", fontsize = 17)
axs[1].set_xlabel("#Querys", fontsize = 17)
axs[2].set_xlabel("#Querys", fontsize = 17)
axs[3].set_xlabel("#Querys", fontsize = 17)
axs[4].set_xlabel("#Querys", fontsize = 17)

axs[1].yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
axs[3].yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
axs[0].set_ylim(0.9, 1.02)
axs[2].set_ylim(0.87, 0.96)
axs[4].set_ylim(0.74, 0.92)
axs[4].yaxis.set_major_formatter(FormatStrFormatter('%.2f'))

handles, labels = axs[1].get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', ncol = 5, fontsize = 17, columnspacing = 3, bbox_to_anchor = (0.5, -0.13))
plt.savefig("distance.pdf", dpi = 600, bbox_inches = 'tight')
plt.show()
```
