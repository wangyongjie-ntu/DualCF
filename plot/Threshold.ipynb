```python
bmap = brewer2mpl.get_map("Set1", "qualitative", 7)                                                                                                                                                                                                                                                                          
colors = bmap.colors
Lcolors = colors[0:5]

for i in range(5):
    color = [i/255. for i in Lcolors[i]]
    Lcolors[i] = color

plt.rcParams['xtick.labelsize']= 17
plt.rcParams['ytick.labelsize']= 17
plt.rcParams["figure.figsize"] = (25, 5)
```

```python
from matplotlib.ticker import FormatStrFormatter
fig, axs = plt.subplots(1, 5)#, sharex=True, sharey = True)
xtick = np.arange(7)
    
# plot the [0] syn_linear
axs[0].plot(xtick, synlinear_CF64, color = Lcolors[2], label = "0.6", marker = "D", markerfacecolor = 'none', markersize = 12)
axs[0].plot(xtick, synlinear_CF73, color = Lcolors[0], label = "0.7", marker = "s", markerfacecolor = 'none', markersize = 12)
axs[0].plot(xtick, synlinear_CF82, color = Lcolors[1], label = "0.8", marker = "o", markerfacecolor = 'none', markersize = 12)
axs[0].plot(xtick, synlinear_CF91, color = Lcolors[3], label = "0.9", marker = "X", markerfacecolor = 'none', markersize = 12)


# plot the [1] syn-nonlinear
axs[1].plot(xtick, synnonlinear_CF64, color = Lcolors[2], marker = "D", markerfacecolor = 'none', markersize = 12)
axs[1].plot(xtick, synnonlinear_CF73, color = Lcolors[0], marker = "s", markerfacecolor = 'none', markersize = 12)
axs[1].plot(xtick, synnonlinear_CF82, color = Lcolors[1], marker = "o", markerfacecolor = 'none', markersize = 12)
axs[1].plot(xtick, synnonlinear_CF91, color = Lcolors[3], marker = "X", markerfacecolor = 'none', markersize = 12)

# plot the [2] GMSC
axs[2].plot(xtick, gmsc_CF64, color = Lcolors[2], marker = "D", markerfacecolor = 'none', markersize = 12)
axs[2].plot(xtick, gmsc_CF73, color = Lcolors[0], marker = "s", markerfacecolor = 'none', markersize = 12)
axs[2].plot(xtick, gmsc_CF82, color = Lcolors[1], marker = "o", markerfacecolor = 'none', markersize = 12)
axs[2].plot(xtick, gmsc_CF91, color = Lcolors[3], marker = "X", markerfacecolor = 'none', markersize = 12)

# plot the [3] Heloc-10
axs[3].plot(xtick, heloc_CF64, color = Lcolors[2], marker = "D", markerfacecolor = 'none', markersize = 12)
axs[3].plot(xtick, heloc_CF73, color = Lcolors[0], marker = "s", markerfacecolor = 'none', markersize = 12)
axs[3].plot(xtick, heloc_CF82, color = Lcolors[1], marker = "o", markerfacecolor = 'none', markersize = 12)
axs[3].plot(xtick, heloc_CF91, color = Lcolors[3], marker = "X", markerfacecolor = 'none', markersize = 12)


# plot [4] Boston
xtick1 = [0, 1, 2, 3, 4, 5]
axs[4].plot(xtick1, boston_CF64, color = Lcolors[2], marker = "D", markerfacecolor = 'none', markersize = 12)
axs[4].plot(xtick1, boston_CF73, color = Lcolors[0], marker = "s", markerfacecolor = 'none', markersize = 12)
axs[4].plot(xtick1, boston_CF82, color = Lcolors[1], marker = "o", markerfacecolor = 'none', markersize = 12)
axs[4].plot(xtick1, boston_CF91, color = Lcolors[3], marker = "X", markerfacecolor = 'none', markersize = 12)


axs[0].set_ylabel(r'Agreement$\rightarrow$', fontsize = 17)

axs[0].set_title(r"Syn-Linear", fontsize = 17)
axs[1].set_title(r"Syn-Nonliear", fontsize = 17)
axs[2].set_title(r"GMSC", fontsize = 17)
axs[3].set_title(r"Heloc-10", fontsize = 17)
axs[4].set_title(r"Boston-Housing", fontsize = 17)

axs[0].set_xticks(xtick)
axs[0].set_xticklabels(synlinear_xaxis)
axs[1].set_xticks(xtick)
axs[1].set_xticklabels(synnonlinear_xaxis)
axs[2].set_xticks(xtick)
axs[2].set_xticklabels(gmsc_xaxis)
axs[3].set_xticks(xtick)
axs[3].set_xticklabels(heloc_xaxis)
axs[4].set_xticks(xtick1)
axs[4].set_xticklabels(boston_xaxis)

axs[0].set_xlabel("#Querys", fontsize = 17)
axs[1].set_xlabel("#Querys", fontsize = 17)
axs[2].set_xlabel("#Querys", fontsize = 17)
axs[3].set_xlabel("#Querys", fontsize = 17)
axs[4].set_xlabel("#Querys", fontsize = 17)

axs[1].yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
axs[0].set_ylim(0.87, 1.01)
axs[1].set_ylim(0.87, 1.01)
axs[2].set_ylim(0.87, 0.96)
#axs[3].set_ylim(0.87, 1.01)
#axs[4].set_ylim(0.87, 0.95)

#axs[0,0].legend(ncol = 5, loc = "lower center",bbox_to_anchor = (0.5, 1.0), fontsize = 14, columnspacing = 1)
handles, labels = axs[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', ncol = 5, fontsize = 17, columnspacing = 3, bbox_to_anchor = (0.5, -0.13))
plt.savefig("threshold.pdf", dpi = 600, bbox_inches = 'tight')
plt.show()
```
