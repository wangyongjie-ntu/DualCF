```python
bmap = brewer2mpl.get_map("Set1", "qualitative", 7)                                                                                                                                                                                                                                                                          
colors = bmap.colors
Lcolors = colors[0:5]

for i in range(5):
    color = [i/255. for i in Lcolors[i]]
    Lcolors[i] = color

plt.rcParams['xtick.labelsize']= 17
plt.rcParams['ytick.labelsize']= 17
plt.rcParams["figure.figsize"] = (15, 5)
```

```python
from matplotlib.ticker import FormatStrFormatter
fig, axs = plt.subplots(1,3)#, sharex=True, sharey = True)
xtick = np.arange(7)
xtick1 = np.arange(6)

# plot the [0, 0] syn_linear
axs[0].plot(xtick, gmsc_prediction, color = Lcolors[0], marker = "s", label = "Prediction Interface", markerfacecolor = 'none', markersize = 12)
axs[0].plot(xtick, gmsc_prediction_core, color = Lcolors[1], marker = "o", label = "Prediction Interface(CoreSet)", markerfacecolor = 'none', markersize = 12)
axs[0].plot(xtick, gmsc_model_ext, color = Lcolors[2], marker = "D", label = "Mode Extraction", markerfacecolor = 'none', markersize = 12)
axs[0].plot(xtick, gmsc_double_cf, color = Lcolors[3], marker = "X", label = "Double CF", markerfacecolor = 'none', markersize = 12)
axs[0].plot(xtick, gmsc_double_cfx, color = Lcolors[4], marker = "P", label = "Double CF + Input", markerfacecolor = 'none', markersize = 12)

axs[1].plot(xtick, heloc_prediction, color = Lcolors[0], marker = "s", markerfacecolor = 'none', markersize = 12)
axs[1].plot(xtick, heloc_prediction_core, color = Lcolors[1], marker = "o", markerfacecolor = 'none', markersize = 12)
axs[1].plot(xtick, heloc_model_ext, color = Lcolors[2], marker = "D", markerfacecolor = 'none', markersize = 12)
axs[1].plot(xtick, heloc_double_cf, color = Lcolors[3], marker = "X", markerfacecolor = 'none', markersize = 12)
axs[1].plot(xtick, heloc_double_cfx, color = Lcolors[4], marker = "P", markerfacecolor = 'none', markersize = 12)

axs[2].plot(xtick1, boston_prediction, color = Lcolors[0], marker = "s", markerfacecolor = 'none', markersize = 12)
axs[2].plot(xtick1, boston_prediction_core, color = Lcolors[1], marker = "o", markerfacecolor = 'none', markersize = 12)
axs[2].plot(xtick1, boston_model_ext, color = Lcolors[2], marker = "D", markerfacecolor = 'none', markersize = 12)
axs[2].plot(xtick1, boston_double_cf, color = Lcolors[3], marker = "X", markerfacecolor = 'none', markersize = 12)
axs[2].plot(xtick1, boston_double_cfx, color = Lcolors[4], marker = "P", markerfacecolor = 'none', markersize = 12)



axs[0].set_ylabel(r'Agreement$\rightarrow$', fontsize = 17)
axs[0].set_title(r"GMSC", fontsize = 17)
axs[1].set_title(r"Heloc-10", fontsize = 17)
axs[2].set_title(r"Boston-Housing", fontsize = 17)

axs[0].set_xticks(xtick)
axs[0].set_xticklabels(gmsc_xaxis)
axs[1].set_xticks(xtick)
axs[1].set_xticklabels(heloc_xaxis)
axs[2].set_xticks(xtick1)
axs[2].set_xticklabels(boston_xaxis)

axs[0].set_xlabel("#Querys", fontsize = 17)
axs[1].set_xlabel("#Querys", fontsize = 17)
axs[2].set_xlabel("#Querys", fontsize = 17)


axs[1].yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
# axs[0].set_ylim(0.9, 1.02)
# axs[2].set_ylim(0.87, 0.96)
# axs[4].set_ylim(0.74, 0.92)

handles, labels = axs[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', ncol = 5, fontsize = 13, columnspacing = 1, bbox_to_anchor = (0.5, -0.1))
plt.savefig("implementation.pdf", dpi = 600, bbox_inches = 'tight')
plt.show()
```
