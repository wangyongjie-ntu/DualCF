```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
%matplotlib inline
```

```python
prediction_mean = [0.734,0.775,0.821,0.861,0.882,0.898,0.912,0.935,0.953]
prediction_std = [0.124,0.106,0.081,0.051,0.024,0.021,0.026,0.032,0.032]

extraction_mean = [0.804,0.835,0.878,0.9,0.911,0.936,0.948,0.933, 0.981]
extraction_std = [0.073,0.058,0.042,0.044,0.045,0.041,0.044,0.033, 0.029]

doublecf_mean = [0.847,0.875,0.9,0.922,0.952,0.979,0.975,0.983,0.986]
doublecf_std = [0.059,0.078,0.067,0.067,0.075,0.03,0.073,0.064, 0.065]
```

```python
X = [3, 5, 10, 20, 40, 80, 160, 320, 611]
```

```python
fig, ax = plt.subplots(2,1)
ax[0].tick_params(axis='both', which='major', labelsize=12)
fig.set_size_inches(5, 10)
X_ = np.arange(len(X))
ax[0].plot(X_, prediction_mean, marker = "o", markersize = 15, markerfacecolor="None", label = "Prediction Interface")
ax[0].plot(X_, extraction_mean, marker = "*", markersize = 15, markerfacecolor="None", label = "Model Extraction")
ax[0].plot(X_, doublecf_mean, marker = "P", markersize = 15, markerfacecolor="None", label = "Double CF")
ax[0].set_xticks(X_)
ax[0].set_xticklabels(X)
#ax[0].set_xlabel('Query number', fontsize = 12)
ax[0].set_ylabel('Agreement', fontsize =12)
ax[0].legend()
ax[0].set_title(r"Syn-Nonlinear Dataset $\uparrow$")

ax[1].plot(X_, prediction_std, marker = "o", markersize = 15, markerfacecolor="None", label = "Prediction Interface")
ax[1].plot(X_, extraction_std, marker = "*", markersize = 15, markerfacecolor="None", label = "Model Extraction")
ax[1].plot(X_, doublecf_std, marker = "P", markersize = 15, markerfacecolor="None", label = "Double CF")
ax[1].set_xticks(X_)
ax[1].set_xticklabels(X)
ax[1].set_xlabel('Query number', fontsize = 12)
ax[1].set_ylabel('Std', fontsize =12)
ax[1].legend()
#ax[1].set_title(r"Syn-Nonlinear Dataset $\downarrow$")

plt.tight_layout()
plt.savefig("Syn_Nonlinear.pdf", dpi = 600)
plt.show()
```
