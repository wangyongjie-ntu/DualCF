```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
%matplotlib inline
```

```python
prediction_mean = [0.642,0.737,0.786,0.848,0.878,0.908,0.925,0.94]
prediction_std = [0.112,0.089,0.062,0.035,0.025,0.014,0.009,0.007]

extraction_mean = [0.831,0.84,0.864,0.885,0.912,0.932,0.944,0.953]
extraction_std = [0.046,0.036,0.03,0.024,0.016,0.011,0.007,0.005]

doublecf_mean = [0.874,0.883,0.893,0.898,0.906,0.915,0.922,0.928]
doublecf_std = [0.022,0.023,0.017,0.02,0.02,0.02,0.016,0.015]

doublecf_input_mean = [0.889,0.894,0.898,0.915,0.928,0.938,0.947]
doublecf_input_std = [0.019,0.018,0.017,0.015,0.012,0.009,0.007]
```

```python
X = [3, 5, 10, 20, 40, 80, 160, 320]
```

```python
fig, ax = plt.subplots(2,1)
ax[0].tick_params(axis='both', which='major', labelsize=12)
fig.set_size_inches(5, 10)
X_ = np.arange(len(X))
idx1 = [1, 2, 3, 4, 5, 6, 7]
idx2 = [1, 2, 3, 4, 5, 6]
ax[0].plot(X_[0:len(prediction_mean)], prediction_mean, marker = "o", markersize = 15, markerfacecolor="None", label = "Prediction Interface")
ax[0].plot(X_[0:len(prediction_mean)], extraction_mean, marker = "*", markersize = 15, markerfacecolor="None", label = "Model Extraction")
ax[0].plot(X_[idx1], doublecf_mean[0:7], marker = "P", markersize = 15, markerfacecolor="None", label = "Double CF")
ax[0].plot(X_[idx1], doublecf_input_mean[0:7], marker = "s", markersize = 15, markerfacecolor="None", label = "Double CF + Input")
ax[0].set_xticks(X_)
ax[0].set_xticklabels(X)
#ax[0].set_xlabel('Query number', fontsize = 12)
ax[0].set_ylabel(r'Agreement $\rightarrow$', fontsize =12)
ax[0].legend()
ax[0].set_title(r"Heloc-10")

ax[1].plot(X_[0:len(prediction_mean)], prediction_std, marker = "o", markersize = 15, markerfacecolor="None", label = "Prediction Interface")
ax[1].plot(X_[0:len(prediction_mean)], extraction_std, marker = "*", markersize = 15, markerfacecolor="None", label = "Model Extraction")
ax[1].plot(X_[idx1], doublecf_std[0:7], marker = "P", markersize = 15, markerfacecolor="None", label = "Double CF")
ax[1].plot(X_[idx1], doublecf_input_std, marker = "s", markersize = 15, markerfacecolor="None", label = "Double CF + Input")
ax[1].set_xticks(X_)
ax[1].set_xticklabels(X)
ax[1].set_xlabel('Query number', fontsize = 12)
ax[1].set_ylabel(r'Std $\leftarrow$', fontsize =12)
ax[1].legend()
#ax[1].set_title(r"Syn-Linear Dataset $\downarrow$")

plt.tight_layout()
plt.savefig("Heloc-10.pdf", dpi = 600)
plt.show()
```
