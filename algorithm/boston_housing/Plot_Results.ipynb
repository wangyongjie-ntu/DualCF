```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
%matplotlib inline
```

```python
prediction_mean = [0.705,0.771,0.86,0.923,0.923,0.957,0.975,0.986,0.99]
prediction_std = [0.165,0.157,0.126,0.076,0.028,0.012,0.026,0.007,0.032]

extraction_mean = [0.804,0.835,0.878,0.9,0.911,0.936,0.948,0.933, 0.981]
extraction_std = [0.073,0.058,0.042,0.044,0.045,0.041,0.044,0.033, 0.029]

doublecf_mean = [0.847,0.875,0.9,0.922,0.952,0.979,0.975,0.983,0.986]
doublecf_std = [0.059,0.078,0.067,0.067,0.075,0.03,0.073,0.064, 0.065]
```

```python
X = [3, 5, 10, 20, 40, 80, 160, 320, 611]
```

```python
fig, ax = plt.subplots()
ax.tick_params(axis='both', which='major', labelsize=12)
fig.set_size_inches(5, 5)
X_ = np.arange(len(X))
ax.plot(X_, prediction_mean, marker = "o", markersize = 15, markerfacecolor="None", label = "Prediction Interface")
ax.plot(X_, extraction_mean, marker = "*", markersize = 15, markerfacecolor="None", label = "Model Extraction")
ax.plot(X_, doublecf_mean, marker = "P", markersize = 15, markerfacecolor="None", label = "Double CF")
ax.set_xticks(X_)
ax.set_xticklabels(X)
ax.set_xlabel('Query number', fontsize = 12)
ax.set_ylabel('Agreement', fontsize =12)
ax.legend()
ax.set_title("Syn-Nonlinear Dataset")
plt.tight_layout()
plt.save_fig("Syn_nonlinear.png")
plt.show()
```
