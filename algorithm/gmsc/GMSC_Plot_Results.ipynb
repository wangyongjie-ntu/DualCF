```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
%matplotlib inline
```

```python
prediction_mean = [0.691,0.789,0.86,0.898,0.929,0.959,0.979,0.986,0.99,0.994]
prediction_std = [0.104,0.074,0.036,0.024,0.019,0.011,0.005,0.003,0.002,0.001]

extraction_mean = [0.864,0.89,0.906,0.927,0.929,0.97,0.985,0.988,0.991,0.994]
extraction_std = [0.026,0.024,0.022,0.018,0.011,0.011,0.003,0.002,0.001,0.001]

doublecf_mean = [0.885,0.923,0.932,0.942,0.953,0.976,0.981,0.984,]
doublecf_std = [0.03,0.018,0.017,0.015,0.009,0.005,0.003,0.002,]
```

```python
X = [5, 10, 20, 40, 80, 160, 320, 640, 1280, 2887]
```

```python
fig, ax = plt.subplots(2,1)
ax[0].tick_params(axis='both', which='major', labelsize=12)
fig.set_size_inches(5, 10)
X_ = np.arange(len(X))
ax[0].plot(X_, prediction_mean, marker = "o", markersize = 15, markerfacecolor="None", label = "Prediction Interface")
ax[0].plot(X_, extraction_mean, marker = "*", markersize = 15, markerfacecolor="None", label = "Model Extraction")
ax[0].plot(X_[0:8], doublecf_mean[0:8], marker = "P", markersize = 15, markerfacecolor="None", label = "Double CF")
ax[0].set_xticks(X_)
ax[0].set_xticklabels(X)
ax[0].set_ylabel('Agreement', fontsize =12)
ax[0].legend()
ax[0].set_title(r"GMSC Dataset $\uparrow$")

ax[1].plot(X_, prediction_std, marker = "o", markersize = 15, markerfacecolor="None", label = "Prediction Interface")
ax[1].plot(X_, extraction_std, marker = "*", markersize = 15, markerfacecolor="None", label = "Model Extraction")
ax[1].plot(X_[0:8], doublecf_std[0:8], marker = "P", markersize = 15, markerfacecolor="None", label = "Double CF")
ax[1].set_xticks(X_)
ax[1].set_xticklabels(X)
ax[1].set_xlabel('Query number', fontsize = 12)
ax[1].set_ylabel('Std', fontsize =12)
ax[1].legend()

plt.tight_layout()
plt.savefig("GMSC.pdf", dpi = 600)
plt.show()
```
