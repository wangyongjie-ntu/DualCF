```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
%matplotlib inline
```

```python
prediction_mean = [0.651,0.717,0.786,0.81,0.844,0.871,0.877]
prediction_std = [0.134,0.12,0.063,0.049,0.027,0.02,0.014]

extraction_mean = [0.767,0.768,0.766,0.798,0.829,0.881,0.897]
extraction_std = [0.07,0.067,0.066,0.051,0.04,0.029,0.018]

doublecf_mean = [0.785,0.797,0.809,0.868,0.909,0.938,0.947]
doublecf_std = [0.074,0.063,0.049,0.038,0.019,0.016,0.013]

doublecf_input_mean = [0.798,0.827,0.818,0.867,0.911,0.939,0.95]
doublecf_input_std = [0.063,0.052,0.055,0.035,0.023,0.014,0.014]
```

```python
X = [3, 5, 10, 20, 40, 80, 126, 160, 252]
```

```python
fig, ax = plt.subplots(2,1)
ax[0].tick_params(axis='both', which='major', labelsize=12)
fig.set_size_inches(5, 10)
X_ = np.arange(len(X))
idx2 = [1, 2, 3, 4, 5, 7, 8]
ax[0].plot(X_[0:len(prediction_mean)], prediction_mean, marker = "o", markersize = 15, markerfacecolor="None", label = "Prediction Interface")
ax[0].plot(X_[0:len(prediction_mean)], extraction_mean, marker = "*", markersize = 15, markerfacecolor="None", label = "Model Extraction")
ax[0].plot(X_[idx2], doublecf_mean, marker = "P", markersize = 15, markerfacecolor="None", label = "Double CF")
ax[0].plot(X_[idx2], doublecf_input_mean, marker = "s", markersize = 15, markerfacecolor="None", label = "Double CF + Input")
ax[0].set_xticks(X_)
ax[0].set_xticklabels(X)
#ax[0].set_xlabel('Query number', fontsize = 12)
ax[0].set_ylabel(r'Agreement $\rightarrow$', fontsize =12)
ax[0].legend()
ax[0].set_title(r"Boston Housing")

ax[1].plot(X_[0:len(prediction_mean)], prediction_std, marker = "o", markersize = 15, markerfacecolor="None", label = "Prediction Interface")
ax[1].plot(X_[0:len(prediction_mean)], extraction_std, marker = "*", markersize = 15, markerfacecolor="None", label = "Model Extraction")
ax[1].plot(X_[idx2], doublecf_std, marker = "P", markersize = 15, markerfacecolor="None", label = "Double CF")
ax[1].plot(X_[idx2], doublecf_input_std, marker = "s", markersize = 15, markerfacecolor="None", label = "Double CF + Input")
ax[1].set_xticks(X_)
ax[1].set_xticklabels(X)
ax[1].set_xlabel('Query number', fontsize = 12)
ax[1].set_ylabel(r'Std $\leftarrow$', fontsize =12)
ax[1].legend()
#ax[1].set_title(r"Syn-Linear Dataset $\downarrow$")

plt.tight_layout()
plt.savefig("Boston Housing.pdf", dpi = 600)
plt.show()
```
