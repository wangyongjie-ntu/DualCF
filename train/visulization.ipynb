```python
import numpy as np
import matplotlib.pyplot as plt
plt.gray()
%matplotlib inline
```

```python
x_query = np.load("../data/MNIST_V1/x_query.npy").squeeze()

x_cf = np.load("./query_pn1.npy").squeeze()
```

```python
plt.figure(figsize=(20, 4))
for i in range(1, 6):
    # display original
    ax = plt.subplot(2, 5, i)
    plt.imshow(x_query[i].reshape(28, 28))
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    # display reconstruction
    ax = plt.subplot(2, 5, i + 5)
    plt.imshow(x_cf[i].reshape(28, 28))
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```
